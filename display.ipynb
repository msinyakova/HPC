```python
#!/bin/python3

import os
import sys
import math
import imageio
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D


def u_analytical(x, y, z):
	t = 20.0 / 2.0 * (h ** 2)
	a = math.pi * (((1.0 + 4.0 + 9.0)/ (Len * Len)) ** 0.5)
	return math.sin(math.pi * x / Len) * \
		   math.sin(2.0 * math.pi * y / Len) * \
		   math.sin(3.0 * math.pi * z / Len) * \
		   math.cos(a * t)


directory1 = input()
# output_file = sys.argv[2]
directory = directory1.split('/')[-2]

Len = 1.0
N = int(directory.split('_')[2][1::])
h = float(Len / (N - 1))
core = int(directory.split('_')[3][1::])

data = np.zeros((N, N, N))
time = np.empty(core)
inaccuracy = np.zeros(20)
```
